## SLICING

- iloc[] => index location (USES INDEX)
- loc => USES INDICES FOR ROWS & NAMES FOR COLUMNS

```python
## iloc[row_num, col_num]
## row_num = 0:3 Get row 1,2,3
## col_num = 0:2 Get Column 1,2
## INDEXING ALWAYS START FROM 0
df2.iloc[0:3,0:2]
```

```python
## LETS GET ROWS => 0:20 & COLUMNS => - 0th Col & 4th Col only
df2.iloc[0:20,[0,4]]
```

```python
## LOC METHOD


## LETS GET FIRST 20 ROWS with 
# `SepalLengthCm` and `PetalLengthCm` column

df2.loc[0:20,("SepalLengthCm","PetalLengthCm")]
```

### CONDITIONAL FETCHING

```python
## GET COLUMNS WHERE `PetalLengthCm` is greater than 6
df2[df2["PetalLengthCm"] > 6]
```

```python
condition1 = df2["PetalLengthCm"] > 3
condition2 = df2["Species"] == 'Iris-virginica'
df2[condition1 & condition2]
```
