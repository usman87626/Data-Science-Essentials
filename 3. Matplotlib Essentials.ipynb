# MATPLOTLIB ESSENTIALS

- Python library used for data visualization
- We can create:
    - Bar-Plots

    - Box-plots
    - Histograms
    - and much more with matplotlib

```python
## IMPORT
from matplotlib import pyplot as plt
## FOR ARRAYS
import numpy as np
```

## Line Plot
- Line plots are generally used to visualize the directional movement of one or more data over time.

```python
x = np.arange(1,11)
x
```

```python
Y = 2 * x
Y
```

```python
plt.plot(x,Y,color="green", linewidth=3, linestyle=":")

plt.title("LINE-PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")

## ADDING GRID
plt.grid(True)

plt.show()
```

```python
## LETS CREATE 2 LINES
x1 = np.arange(0,11)
y1 = x1 * 2
y2 = x1 * 3
```

```python
# LINE PLOT
plt.plot(x1,y1,color="green", linewidth=3, linestyle=":")
plt.plot(x1,y2,color="red", linewidth=3, linestyle="-")

plt.title("LINE-PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")

## ADDING GRID
plt.grid(True)

plt.show()
```

## Bar Plot

- Used for categorical data instead of continuos data.
- A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.
- The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a column chart.

```python
student = {
    "Usman": 30,
    "Ali": 50,
    "Ahmad": 70
}

names = list(student.keys())
marks = list(student.values())
print(names)
print(marks)
```

```python
## PLOT BAR-PLOT
plt.bar(names, marks,color="red")

plt.title("Student Marks")

plt.xlabel("Names")
plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

```python
## PLOT BAR-PLOT
plt.barh(names, marks,color="red")

plt.title("Student Marks")

plt.xlabel("Marks")
plt.ylabel("Name")

plt.grid(True)

plt.show()
```

### Scatter Plots
- Also known as scatterplot, scatter graph, scatter chart, scattergram, or scatter diagram
- Scatter plots are used to observe and show relationships between two numeric variables
- Type of plot using Cartesian coordinates to display values for typically two variables for a set of data.
-  The data are displayed as a collection of points:
    - Each having the value of one variable determining the position on the horizontal axis 
    - & the value of the other variable determining the position on the vertical axis.

```python
x = [1,2,3,9,1,6]

y = [1,4,2,6,4,9]

plt.scatter(x,y)

plt.grid(True)

plt.show()
```

```python
## TWO THEMES ON SINGLE SCATTER-PLOT
x = [1,3,5,7,9,11]

y1 = [2,4,6,8,10,12]

y2 = [3,5,7,9,11,13]

plt.scatter(x,y1)

plt.scatter(x,y2,color="red")

plt.grid(True)

plt.show()
```

## HISTOGRAM
- Helps us to Understand distribution of continuos columns
- Let's see it using `iris.csv`
- `Bar Plot` is different from `Histogram`in a sense that:
    - Bar plot is used for categorical data
    - Histogram is used for continuos data
    

```python
import pandas as pd

df = pd.read_csv("data/iris.csv")
df.head()
```

```python
plt.hist(df['SepalLengthCm'])

plt.show()
```

```python
# WE CAN EXPAND IT BY INCREASING NO. Of Bins
plt.hist(df['SepalLengthCm'],bins=20)

plt.show()
```

```python
# WE CAN EXPAND IT BY INCREASING NO. Of Bins
plt.hist(df['PetalWidthCm'],bins=20)

plt.show()
```

## Box Plot
- Used to understand the distribution of how does one continuos variable change with respect to a categorical value

```python
## LET'S SEE HOW DOES SEPAL LENGTH VARIES WITH SPECIES
df.boxplot(column='SepalLengthCm',by='Species')
```

### INTERPRETATIONS FROM ABOVE BOX-PLOT
- If the species of the flower is `Iris-setosa`, the median sepal length of the flower would be around 5
- If the species of the flower is `Iris-versicolor`, the median sepal length would be around 5.9
- If the species of the flower is `Iris-virginica`, the median sepal length would be around 6.5
<hr>
- `Iris-setosa` has the minimum sepal length while `Iris-virginica` has the maximum sepal length.

```python
## LET'S SEE HOW DOES PETAL LENGTH VARIES WITH SPECIES
df.boxplot(column='PetalWidthCm',by='Species')
```

### INTERPRETATIONS FROM ABOVE BOX-PLOT
- If the species of the flower is `Iris-setosa`, the median petal width of the flower would be around 0.3
- If the species of the flower is `Iris-versicolor`, the median petal width would be around 1.2
- If the species of the flower is `Iris-virginica`, the median petal width would be around 2.0
<hr>
- `Iris-setosa` has the minimum petal width while `Iris-virginica` has the maximum petal width.

```python
## LETS DO IT MORE BEAUTIFULLY USING SEABORN
import seaborn as sns

sns.boxplot(x=df['Species'],y=df['SepalLengthCm'])
```

## Pie Chart
- A pie chart (or a circle chart) is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.
- In a pie chart, the arc length of each slice (and consequently its central angle and area), is proportional to the quantity it represents.
- While it is named for its resemblance to a pie which has been sliced, there are variations on the way it can be presented.

```python
fruits = ["Apple","Orange","Banana","Mango"]
cost = [76,45,90,85]
```

```python
plt.pie(cost, labels=fruits)
plt.show()
```

```python
plt.pie(cost, labels=fruits, autopct='%0.2f%%',shadow=True)
plt.show()
```

```python
#INCREASE FIGURE SIZE
plt.figure(figsize=(10,10))
plt.pie(cost, labels=fruits, autopct='%0.2f%%',shadow=True)
plt.show()
```
